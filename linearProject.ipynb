```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='white', color_codes=True)
```

```python
SeoulBike = pd.read_csv("/content/SeoulBikeData.csv")
SeoulBike.head()
```

```python
print("Dimension of SeoulBike dataset", SeoulBike.shape)
```

```python
data = pd.concat([SeoulBike], ignore_index = True)
```

```python
data.head()
```

```python
df = data.select_dtypes(include=['float64','int64'])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Identifing missing Values
miss_val = df.isna().sum()
miss_val
```

```python
miss_val[miss_val>0].sort_values(ascending = False)
```

```python
df_SeoulBike = df[:1460]
```

```python
#Checking for outliers using boxplot

def plot_box( col_name):
  plt.figure(figsize = (5,5))
  sns.boxplot(df_SeoulBike[col_name])
  plt.show()

plot_box('Dew point temperature(C)')
```

```python
df_SeoulBike.columns
```

```python
plot_box('Wind speed (m/s)')

# REmoving the outliers


q1, q2, q3 = np.percentile(df_SeoulBike['Wind speed (m/s)'], [25,50,75])

int_quart_range = q3 - q1
df_SeoulBike = df_SeoulBike[df_SeoulBike['Wind speed (m/s)'] > (q1 - (int_quart_range * 1.5))]
df_SeoulBike = df_SeoulBike[df_SeoulBike['Wind speed (m/s)'] < (q3 + (int_quart_range * 1.5))]

plot_box('Wind speed (m/s)')
```

```python
df_SeoulBike.shape
```

```python
data_corr = df_SeoulBike.corr()
data_corr
```

```python
corr = df_SeoulBike.corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr, annot=True, cmap='cubehelix_r',square=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(df_SeoulBike, test_size=0.25)
```

```python
SeoulBike_x = train.drop(['Rented Bike Count'], axis =1)
SeoulBike_y = train['Rented Bike Count'] 

val_x = val.drop(['Rented Bike Count'], axis = 1)
val_y = val['Rented Bike Count']
```

```python
print('Dimension of SeoulBike_x dataframe:',SeoulBike_x.shape)
print('Dimension of SeoulBike_y dataframe:',SeoulBike_y.shape)
print('Dimension of val_x dataframe:',val_x.shape)
print('Dimension of val_y dataframe:',val_y.shape)
```

```python
df_SeoulBike.columns
```

```python
from sklearn.linear_model import LinearRegression
ln_reg =  LinearRegression()
ln_reg.fit(SeoulBike_x, SeoulBike_y)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

df_pred = ln_reg.predict(val_x)
print('Mean Absolute Error of linear regression:',mean_absolute_error(df_pred, val_y))
print('Mean Square Error of linear regression:', mean_squared_error(df_pred, val_y))
print('R_Squared Score of linear regression:', r2_score(val_y, df_pred))
```
