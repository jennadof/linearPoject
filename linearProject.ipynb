```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='white', color_codes=True)
```

```python
data.drop(['Date'],axis = 1, inplace = True)
```

```python
df = data
```

```python
for cols in df:
  dummies = pd.get_dummies(data[cols], drop_first = True)
  df = pd.concat([df, dummies ], axis ="columns")
df
```

```python
df['Seasons'].value_counts()
```

```python
#Checking for outliers using boxplot

def plot_box( col_name):
  plt.figure(figsize = (5,5))
  sns.boxplot(df[col_name])
  plt.show()

plot_box('Dew point temperature(C)')
```

```python
plot_box('Wind speed (m/s)')

# REmoving the outliers


q1, q2, q3 = np.percentile(df['Wind speed (m/s)'], [25,50,75])

int_quart_range = q3 - q1
df = df[df['Wind speed (m/s)'] > (q1 - (int_quart_range * 1.5))]
df = df[df['Wind speed (m/s)'] < (q3 + (int_quart_range * 1.5))]

plot_box('Wind speed (m/s)')
```

```python
x = df.drop(["Rented Bike Count",'Holiday','Functioning Day','Seasons'], axis = 1)
```

```python
y = df["Rented Bike Count"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
ln_reg =  LinearRegression()
ln_reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

df_pred = ln_reg.predict(x_test)
print('Mean Absolute Error of linear regression:',mean_absolute_error(df_pred, y_test))
print('Mean Square Error of linear regression:', mean_squared_error(df_pred,y_test))
print('R_Squared Score of linear regression:', r2_score(df_pred, y_test))
```
